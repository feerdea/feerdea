```python
# 1. Importing libraries:

import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# 2. Reading a CSV file:

data = pd.read_csv("countries.csv")
data.head()
```

```python
# 3. Describing the data:

data.describe()
```

```python
# 4. Grouping the data by a column:

grouped_data = data.groupby("year").mean()
grouped_data
```

```python
# 5. Creating a histogram:

plt.hist(data["population"])
plt.show()
```

```python
# 6. Creating a scatter plot:

plt.scatter(data["year"], data["population"])
plt.xlabel("year")
plt.ylabel("population")
plt.show()
```

```python
# 7. Creating a line plot:

plt.plot(data["year"], data["population"])
plt.xlabel("year")
plt.ylabel("population")
plt.show()
```

```python
# 8. Seaborn library:

import seaborn as sns
sns.set(style="ticks")

# Creating a boxplot
sns.boxplot(x="year", y="population", data=data)
plt.show()

# Creating a lineplot
sns.lineplot(x="year", y="population", data=data)
plt.show()

# Creating a heatmap
correlation = data.corr()
sns.heatmap(correlation, annot=True)
plt.show()
```

```python
# 9. Plotly library:

import plotly.express as px

# Creating a scatter plot
fig = px.scatter(data, x='year', y='population', color='population')
fig.show()

# Creating a bar chart
fig = px.bar(data, x='year', y='population', color='population')
fig.show()
```

```python
# 10. Bokeh library:

from bokeh.plotting import figure, show
from bokeh.io import output_notebook

output_notebook()

# Creating a line plot
p = figure(title="Line Plot", x_axis_label='year', y_axis_label='population')
p.line(data["year"], data["population"])
show(p)

# Creating a scatter plot
p = figure(title="Scatter Plot", x_axis_label='year', y_axis_label='population')
p.scatter(data["year"], data["population"])
show(p)
```
